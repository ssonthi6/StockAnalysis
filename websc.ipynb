```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
import csv
```

```python
# Set the URL of the Yahoo Finance options page
url = "https://finance.yahoo.com/quote/AAPL/options?p=AAPL"

# Initialize the webdriver
driver = webdriver.Chrome()
driver.get(url)

# Wait for the pop-up to appear and then close it
wait = WebDriverWait(driver, 10)
popup = driver.find_element(By.XPATH, "//*[@id=\"myLightboxContainer\"]/section/button[2]")
popup.click()

# Wait for the options table to load
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.ID, "mrt-node-Col1-1-OptionContracts")))

# Find the options table
options_table = driver.find_element(By.ID, "mrt-node-Col1-1-OptionContracts")

# Get the rows of the options table
rows = options_table.find_elements(By.TAG_NAME, "tr")

# Iterate through the rows and print the data
for row in rows:
    cells = row.find_elements(By.TAG_NAME, "td")
    data = [cell.text for cell in cells]
    print(data)

# Close the webdriver
driver.close()
```

```python
# Set the URL of the Yahoo Finance options page
url = "https://finance.yahoo.com/quote/AAPL/options?p=AAPL"

chrome_options = Options()
chrome_options.add_argument("--headless")

# Initialize the webdriver
driver = webdriver.Chrome(options=chrome_options)
driver.get(url)

# Wait for the pop-up to appear and then close it
try:
    wait = WebDriverWait(driver, 10)
    popup = driver.find_element(By.XPATH, "//*[@id=\"myLightboxContainer\"]/section/button[2]")
    popup.click()
except:
    print("x")
    
# Wait for the options table to load
wait = WebDriverWait(driver, 10)
wait.until(EC.presence_of_element_located((By.XPATH, "//*[@id=\"Col1-1-OptionContracts-Proxy\"]/section/div/div[1]/select")))

# Find the options table
date_tables = driver.find_element(By.XPATH, "//*[@id=\"Col1-1-OptionContracts-Proxy\"]/section/div/div[1]/select")

# Get the rows of the options table
rows = date_tables.find_elements(By.TAG_NAME, "option")

dates = []
# Iterate through the rows and print the data
for row in rows:
    print(row.get_attribute("value"))
    dates.append(row.get_attribute("value"))

# Close the webdriver
driver.close()
```

```python
chrome_options = Options()
chrome_options.add_argument("--headless")

# Initialize the webdriver
driver = webdriver.Chrome(options=chrome_options)

for d in dates:
    # Set the URL of the Yahoo Finance options page
    url = "https://finance.yahoo.com/quote/AAPL/options?p=AAPL&date=" + str(d)

    # Initialize the webdriver
    driver.get(url)

    # Wait for the pop-up to appear and then close it
    try:
        wait = WebDriverWait(driver, 10)
        popup = driver.find_element(By.XPATH, "//*[@id=\"myLightboxContainer\"]/section/button[2]")
        popup.click()
    except:
        print("x")

    # Wait for the options table to load
    wait = WebDriverWait(driver, 10)
    wait.until(EC.presence_of_element_located((By.ID, "mrt-node-Col1-1-OptionContracts")))

    # Find the options table
    options_table = driver.find_element(By.ID, "mrt-node-Col1-1-OptionContracts")

    # Get the rows of the options table
    rows = options_table.find_elements(By.TAG_NAME, "tr")

    # Iterate through the rows and print the data
    tot_data = []
    for row in rows:
        cells = row.find_elements(By.TAG_NAME, "td")
        data = [cell.text for cell in cells]
        tot_data.append(data)
        
    with open(str(d)+'_t.csv', 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerows(tot_data)
    print("written " + str(d))

# Close the webdriver
driver.close()
```
